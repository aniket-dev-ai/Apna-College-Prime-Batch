### `match`
Short example and output below.

```python
# match example (requires Python 3.10+)
def show(x):
    match x:
        case 0:
            return 'zero'
        case [a,b]:
            return f'list {a},{b}'
        case _:
            return 'other'
print(show(0))
print(show([1,2]))
print(show('hi'))
```

### `nonlocal`
Short example and output below.

```python
# nonlocal example
def outer():
    x=0
    def inner():
        nonlocal x
        x+=2
        return x
    inner()
    return x
print(outer())
```
