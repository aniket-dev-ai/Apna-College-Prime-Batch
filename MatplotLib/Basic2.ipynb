```python
import matplotlib.pyplot as plt

people = [
    "Person A", "Person B", "Person C", "Person D", "Person E",
    "Person F", "Person G", "Person H", "Person I", "Person J"
]

age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]

blood_pressure = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]
```

```python
plt.scatter(age,blood_pressure)
plt.title("Age vs BP")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")

plt.grid()

```

```python
colors = ["green" if x < 135 else "red" for x in blood_pressure]
plt.scatter(age,blood_pressure , color=colors)
plt.title("Age vs BP")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")


plt.grid()

```

```python
colors = ["green" if x < 135 else "red" for x in blood_pressure]

plt.scatter(age,blood_pressure , color=colors , sizes = blood_pressure)
plt.title("Age vs BP")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")


plt.grid()

```
