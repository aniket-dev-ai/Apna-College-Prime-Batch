```python
# Real data

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

oscar_years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M


#plot

plt.plot(oscar_years , oscar_revenue)
plt.title("Oscar Movies Revenue in each year")
plt.xlabel("Oscar Years")
plt.ylabel("Oscar Revenue (in $M)")
```

```python
# Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_years = [2008, 2009, 2010, 2011, 2012]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

plt.plot(oscar_years , oscar_revenue , label="Oscar Movies")
plt.plot(non_oscar_years , non_oscar_revenue , label="Non-Oscar Movies")

plt.title("Oscar vs Non Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")

plt.legend()
```
