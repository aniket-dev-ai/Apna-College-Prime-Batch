```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data.csv")

df
```

```python
df.isnull()
df.isna()
```

```python
df.isnull().sum()
```

```python
df_pure_rows = df.dropna()
df_pure_rows
```

```python
df_pure_cols = df.dropna(axis = 1)
df_pure_cols
```

```python
age_mean = df["age"].mean()

df_aged = df["age"].fillna(age_mean)

df["age"] = df_aged

df
```

```python
income_mean = df["income"].mean()

df_income = df["income"].fillna(income_mean)

df["income"] = df_income

df
```
