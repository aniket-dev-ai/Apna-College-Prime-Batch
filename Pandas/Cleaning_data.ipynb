```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data.csv")

df
```

```python
df.isnull()
df.isna()
```

```python
df.isnull().sum()
```

```python
df_pure_rows = df.dropna()
df_pure_rows
```

```python
df_pure_cols = df.dropna(axis = 1)
df_pure_cols
```

```python
age_mean = df["age"].mean()

df_aged = df["age"].fillna(age_mean)

df["age"] = df_aged

df
```

```python
income_mean = df["income"].mean()

df_income = df["income"].fillna(income_mean)

df["income"] = df_income

df
```

```python
df.ffill()
```

```python
df.duplicated()
```

```python
df = df.drop_duplicates()

df
```

```python
df.duplicated()
```

```python
df["tax_slb"] = df["income"].apply(lambda x : "20%" if x >= 50000 else "10%")

df
```

```python

gender_map = {"Male":"M" , "Female":"F" , "Unknown" : "U" ,"NaN" : "U"}

df["gender"] = df["gender"].map(gender_map)

df
```

```python
df = df.assign(new_income = df["income"] * 1.3)
df
```

```python
df = df.rename(columns={"income":"salary"})
```

```python
df["salary"].sort_values()
```

```python
df.sort_values(["salary","age"])
```

```python
df.sort_index()
```

```python
df = df.assign(salary_rank = df["salary"].rank())
```

```python
df.sort_values("salary_rank", ascending = False)
```

```python

df.to_csv("changed_data.csv")
```
