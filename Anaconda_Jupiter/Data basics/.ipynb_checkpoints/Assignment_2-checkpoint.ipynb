## Get meaningful insights from data

```python
import json
```

```python
def load_data(data):
    with open(data , "r") as f:
        dataa = json.load(f)
    return dataa
```

```python
## Loading data
data = load_data("store_data.json")
print(data)
```

```python
def get_insights(data):

    tot_rating = 0
    for user in data:
        tot_rating += int(user["rating"])

    print(f"avg rating = {tot_rating/len(data)}")
```

```python
get_insights(data)
```
