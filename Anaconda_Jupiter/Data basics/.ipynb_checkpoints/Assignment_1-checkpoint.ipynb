```python
import json
```

## Load the data

```python
def load_data(filename):
    with open(filename , "r") as f:
        data = json.load(f)

    return data
```

```python
data = load_data("store_data.json")

print(data)
```

```python
def clean_data(data):
    text_to_num = {
        "one": 1, "two": 2, "three": 3, "four": 4, "five": 5,
        "six": 6, "seven": 7, "eight": 8, "nine": 9, "ten": 10
    }

    cleaned_users = []
    seen_users = set()   

    for user in data:
        
        raw_rating = str(user.get("rating", "")).strip().lower()

        if raw_rating.isdigit():
            user["rating"] = int(raw_rating)
        elif raw_rating in text_to_num:
            user["rating"] = text_to_num[raw_rating]
        else:
            user["rating"] = None  

         
        raw_age = user.get("age")
        user["age"] = int(raw_age) if str(raw_age).isdigit() else None

        
        key = f'{user.get("first","").strip().lower()}_{user.get("last","").strip().lower()}'

        if key in seen_users:
            continue

        seen_users.add(key)
        cleaned_users.append(user)

    return cleaned_users
```

```python
final_data = clean_data(data)

print(final_data)
```
